#### Problema 1: $x + 8 = 15$
#### Planteamiento: Un número sumado a 8 da 15.
#### Despeje: El 8 está sumando, pasa al otro lado restando.$x = 15 - 8$
#### Resultado: $x = 7$

```python
# Problema 1: X + 8 = 15
# Despeje: X = 15 - 8
#constante_derecha = 15
#valor_sumado = 8
constante_derecha = int(input("Ingresa el valor de la constante a la derecha: "))
valor_sumado = int(input("Ingresa el valor a sumar: "))

x = constante_derecha - valor_sumado
print(f"El resultado de X en el Problema 1 es: {x}")
```

#### Problema 3: $x + 4 = 21$
#### Planteamiento: Un número sumado a 4 es igual a 21.
#### Despeje: El 4 pasa restando al lado derecho.$x = 21 - 4$
#### Resultado: $x = 17$

```python
# Problema 3: X + 4 = 21
# Despeje: X = 21 - 4
#total = 21
#numero = 4
total = int(input("Ingresa el valor de la constante a la derecha: "))
numero = int(input("Ingresa el valor a restar: "))
x = total - numero
print(f"El resultado de X en el Problema 3 es: {x}")
```

#### Problema 5: $6a - 2a + a = 20$
#### Planteamiento: Varias veces una misma variable. Primero agrupamos.
#### Simplificación: $6a - 2a = 4a$. Luego $4a + 1a = 5a$.$5a = 20$
####Despeje: El 5 multiplica a la $a$, así que pasa dividiendo.$a = 20 / 5$
#### Resultado: $a = 4$

```python
# Problema 5: 6A - 2A + A = 20
# Simplificación de términos: (6 - 2 + 1)A = 5A
# Ecuación: 5A = 20 -> A = 20 / 5
#resultado_ecuacion = 20
#coeficiente_total = 6 - 2 + 1 # Suma de los coeficientes de A
resultado_ecuacion = int(input("Ingresa el valor de la constante a la derecha: "))
coeficiente_total = int(input("Ingresa el valor del coeficiente de A: "))
a = resultado_ecuacion / coeficiente_total
print(f"El resultado de A en el Problema 5 es: {int(a)}")
```

#### Problema 7: $4x + 3 = 2x + 11$
#### Planteamiento: Hay variables y números en ambos lados.
#### Agrupación: Pasamos las $x$ a la izquierda y los números a la derecha.$4x - 2x = 11 - 3$
#### Simplificación: $2x = 8$Despeje: $x = 8 / 2$
#### Resultado: $x = 4$

```python
# Problema 7: 4X + 3 = 2X + 11
# Agrupamos X a la izquierda y números a la derecha: 4X - 2X = 11 - 3
# Simplificación: 2X = 8 -> X = 8 / 2
#dif_x = 4 - 2     # Coeficientes de X
#dif_num = 11 - 3  # Constantes
dif_x = int(input("Ingresa el valor del coeficiente de X: "))
dif_num = int(input("Ingresa el valor de la constante: "))
x = dif_num / dif_x
print(f"El resultado de X en el Problema 7 es: {int(x)}")
```

#### Problema 9: $9m + 4 = 5m + 20$

#### Agrupación: $9m - 5m = 20 - 4$
#### Simplificación: $4m = 16$
#### Despeje: $m = 16 / 4$

```python
# Problema 9: 9M + 4 = 5M + 20
# Agrupamos: 9M - 5M = 20 - 4
# Simplificación: 4M = 16 -> M = 16 / 4
#m_izquierda = 9
#m_derecha = 5
#num_derecha = 20
#num_izquierda = 4
m_izquierda = int(input("Ingresa el valor del coeficiente de MIZQ: "))
m_derecha = int(input("Ingresa el valor del coeficiente de MDER: "))
num_derecha = int(input("Ingresa el valor de la constanteDER: "))
num_izquierda =int(input("Ingresa el valor de la constanteIZQ: "))

m = (num_derecha - num_izquierda) / (m_izquierda - m_derecha)
print(f"El resultado de M en el Problema 9 es: {int(m)}")
```

#### Problema 11: $6a + 5 = 2a + 21$
#### Agrupación: Movemos $2a$ a la izquierda (restando) y $5$ a la derecha (restando).$6a - 2a = 21 - 5$
#### Simplificación: $4a = 16$
#### Despeje: $a = 16 / 4$
#### Resultado: $a = 4$

```python
# Problema 11: 6A + 5 = 2A + 21
# Agrupamos: 6A - 2A = 21 - 5
# Simplificación: 4A = 16 -> A = 16 / 4
#coef_a1 = 6
#coef_a2 = 2
#num1 = 21
#num2 = 5
coef_a1 = int(input("Ingresa el valor del coeficiente de A1: "))
coef_a2 = int(input("Ingresa el valor del coeficiente de A2: "))
num1 = int(input("Ingresa el valor de la constante1: "))
num2 = int(input("Ingresa el valor de la constante2: "))

a = (num1 - num2) / (coef_a1 - coef_a2)
print(f"El resultado de A en el Problema 11 es: {int(a)}")
```
